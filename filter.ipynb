```python
import pandas as pd
```

```python
df = pd.read_excel('CLASS IXB STUDENTS DATA.xlsx', sheet_name='IXB2')
```

```python
df.columns
```

```python
df.drop('Email Address', axis=1, inplace=True)
df.drop('Nationality', axis=1, inplace=True)
df.drop('Sr No', axis=1, inplace=True)
df.fillna("", inplace=True)
```

```python
df = df.rename(columns={'Admn No': 'dummy'})
df = df.rename(columns={'Name': 'Admn No'})
df = df.rename(columns={'dummy': 'Name'})
df['Name'], df['Admn No'] = df['Admn No'], df['Name']
```

```python
df.head()
```

```python
from docx import Document
```

```python
def tableMaker(doc: Document(), name: str):
    table = doc.add_table(rows=7, cols=7, style='Table Grid')
    cells = table.rows[0].cells
    cells[0].text = 'Exam'
    cells[1].text = 'Hindi'
    cells[2].text = 'English'
    cells[3].text = 'Maths'
    cells[4].text = 'Science'
    cells[5].text = 'Social Science'
    cells[6].text = 'IT'

    table.rows[1].cells[0].text = 'UT1'
    table.rows[2].cells[0].text = 'UT2'
    table.rows[3].cells[0].text = 'HYE'
    table.rows[4].cells[0].text = 'UT3'
    table.rows[5].cells[0].text = 'UT4'
    table.rows[6].cells[0].text = 'AE'

    doc.add_page_break()
```

```python
doc = Document()

for index, row in df.iterrows():

    # Add a new paragraph to the Word document for each column in the row
    for column in df.columns:
        value = row[column]
        input = f"{column} - {value}"
        doc.add_paragraph(str(input))

    name = str(row[df.columns[0]]) + '.docx'

    tableMaker(doc=doc, name=name)


doc.save('ALL_NAMES.docx')

    
```
